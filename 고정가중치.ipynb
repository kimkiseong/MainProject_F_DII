```python
main["output_index"] = pd.to_numeric(main["output_index"], errors="coerce")
main["labor_productivity_index"] = pd.to_numeric(main["labor_productivity_index"], errors="coerce")
```

```python
group_columns = {
    "공정": [  # 산업 생산성
        'output_index',                # 산출량지수
        'labor_productivity_index',   # 1인당 노동생산성 지수
        'avg_income'                  # 근로자 임금 (변화율 없지만, 대푯값으로 사용 가능)
    ],
    "경쟁": [  # 산업 경쟁력
        'avg_age',                    # 평균 나이 (시장 비고령률 → 나이로 대체)
        'n_ent'                       # 기업 수 (기업 증가율 대신, 시계열로 증감률 계산 가능)
    ],
    "시장": [  # 시장 규모
        'employment_thousand',       # 취업자 수
        'avg_ent_tdwrkr'             # 종사자 수 (기업당 기준이지만 참고 가능)
    ],
    "기업": [  # 산업 이익
        'avg_ent_sls',               # 기업당 매출
        'nominal_value_added'        # 부가가치율 (실제는 총액이지만 활용 가능)
    ]
}
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def prepare_group_scores_with_pca_info(df, group_columns):
    group_scores = {}
    group_pca_info = {}

    for group_name, columns in group_columns.items():
        # 1. 해당 그룹 컬럼만 추출
        temp_df = df[columns].dropna()

        # 2. 정규화
        scaler = StandardScaler()
        scaled = scaler.fit_transform(temp_df)

        # 3. PCA
        pca = PCA(n_components=1)
        score = pca.fit_transform(scaled)

        # 4. 결과 저장
        group_scores[group_name] = pd.Series(score.flatten(), index=temp_df.index)
        group_pca_info[group_name] = {
            "explained_variance_ratio": pca.explained_variance_ratio_,
            "components": pca.components_,
            "pca_model": pca,  # 필요 시 전체 PCA 정보 접근 가능
            "columns": columns
        }

    return group_scores, group_pca_info
```

```python
# 새로운 함수 사용
group_scores, group_pca_info = prepare_group_scores_with_pca_info(main, group_columns)

# 각 그룹별 score를 main에 추가
for group, score_series in group_scores.items():
    main.loc[score_series.index, f'{group}_score'] = score_series
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def prepare_group_scores_with_pca_info(df, group_columns):
    group_scores = {}
    group_pca_info = {}

    for group_name, columns in group_columns.items():
        # 1. 해당 그룹 컬럼만 추출
        temp_df = df[columns].dropna()

        # 2. 정규화
        scaler = StandardScaler()
        scaled = scaler.fit_transform(temp_df)

        # 3. PCA
        pca = PCA(n_components=1)
        score = pca.fit_transform(scaled)

        # 4. 결과 저장
        group_scores[group_name] = pd.Series(score.flatten(), index=temp_df.index)
        group_pca_info[group_name] = {
            "explained_variance_ratio": pca.explained_variance_ratio_,
            "components": pca.components_,
            "pca_model": pca,  # 필요 시 전체 PCA 정보 접근 가능
            "columns": columns
        }

    return group_scores, group_pca_info
```

# 시각화

```python
# 균등 가중치 기반
main["decline_index_uniform"] = (
    0.25 * main["공정_score"] +
    0.25 * main["경쟁_score"] +
    0.25 * main["시장_score"] +
    0.25 * main["기업_score"]
)

# 설명분산 기반
explained_variance = {
    '공정': 0.6236,
    '경쟁': 0.5503,
    '시장': 0.5144,
    '기업': 0.5087
}
total = sum(explained_variance.values())
weights = {k: v / total for k, v in explained_variance.items()}

main["decline_index_pca"] = (
    weights['공정'] * main['공정_score'] +
    weights['경쟁'] * main['경쟁_score'] +
    weights['시장'] * main['시장_score'] +
    weights['기업'] * main['기업_score']
)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=main["decline_index_uniform"],
    name='균등 가중치',
    opacity=0.6
))
fig.add_trace(go.Histogram(
    x=main["decline_index_pca"],
    name='설명분산 가중치',
    opacity=0.6
))

fig.update_layout(
    barmode='overlay',
    title="사양산업지수 분포 비교",
    xaxis_title="사양산업지수",
    yaxis_title="빈도수",
    template="plotly_white"
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter(
    main,
    x="decline_index_uniform",
    y="decline_index_pca",
    title="사양산업지수: 균등 vs 설명분산 가중치",
    labels={
        "decline_index_uniform": "균등 가중치 지수",
        "decline_index_pca": "설명분산 가중치 지수"
    }
)

fig.add_shape(
    type="line",
    x0=main["decline_index_uniform"].min(),
    y0=main["decline_index_uniform"].min(),
    x1=main["decline_index_uniform"].max(),
    y1=main["decline_index_uniform"].max(),
    line=dict(color="red", dash="dash"),
    name="y = x"
)

fig.update_traces(marker=dict(opacity=0.4))
fig.update_layout(template="plotly_white")
fig.show()
```
