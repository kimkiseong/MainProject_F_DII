```python
import numpy as np

# industry_code 붙인 원본에서 melt
value_vars = [col for col in df_산출량지수.columns if isinstance(col, float) or isinstance(col, int)]

df_산출량지수_long = df_산출량지수.melt(
    id_vars=["한국표준산업분류(KSIC 9차)", "industry_code"],
    value_vars=value_vars,
    var_name="period",
    value_name="산출량지수"
).rename(columns={"한국표준산업분류(KSIC 9차)": "industry_name"})

def convert_period_to_base_year(period):
    # float 처리 보장
    period = float(period)
    year = int(period)
    decimal = round(period - year, 2)  # 소수점 이하 정밀도 보정

    # 정수 연도인 경우 → 전년도 4분기로 간주
    if decimal == 0:
        return (year - 1) * 10 + 4
    # 분기별 매핑
    quarter_map = {
        0.25: 1,
        0.5: 2,
        0.75: 3
    }
    quarter = quarter_map.get(decimal, 0)  # 잘못된 값이면 0 반환
    return year * 10 + quarter

df_산출량지수_long['base_year'] = df_산출량지수_long['period'].apply(convert_period_to_base_year)
```

```python
def convert_half_year_to_base_year(period_str):
    """
    예: '2018. 1/2' → [20181, 20182]
        '2018. 2/2' → [20183, 20184]
    """
    try:
        parts = period_str.replace(" ", "").split(".")
        year = int(parts[0])
        half = parts[1]

        if half == '1/2':
            return [year * 10 + 1, year * 10 + 2]
        elif half == '2/2':
            return [year * 10 + 3, year * 10 + 4]
        else:
            return []
    except Exception as e:
        print(f"[⚠️ 오류] '{period_str}' 파싱 실패: {e}")
        return []

# 확장
expanded_rows = []
for _, row in df_취업자_집계.iterrows():
    base_years = convert_half_year_to_base_year(row['period'])
    for base_year in base_years:
        expanded_rows.append({
            '지역': row['행정구역별'],
            'industry_name': row['항목'],
            'base_year': base_year,
            '취업자수_천명': row['취업자수_천명']
        })

df_취업자_expanded = pd.DataFrame(expanded_rows)




```

```python
# 1. 기준값 추출 (68번 산업)
ref_df = df_merged[
    (df_merged['industry_code'] == 68) &
    df_merged['output_index'].notna() &
    df_merged['labor_productivity_index'].notna()
][['sido', 'base_year_x', 'output_index', 'labor_productivity_index']].copy()

ref_df = ref_df.rename(columns={
    'output_index': 'ref_output_index',
    'labor_productivity_index': 'ref_labor_productivity_index'
})

# 2. 76번 산업 중 결측행 추출
target_idx = df_merged[
    (df_merged['industry_code'] == 76) &
    (df_merged['output_index'].isna() | df_merged['labor_productivity_index'].isna())
].index

target_df = df_merged.loc[target_idx].copy()
target_df['original_index'] = target_df.index  # 인덱스 보존

# 3. merge 후 ref값 가져오기
target_df = target_df.merge(ref_df, on=['sido', 'base_year_x'], how='left')

# 4. 다시 인덱스를 원래 위치로
target_df = target_df.set_index('original_index')

# 5. 원본에 반영 (이제 인덱스 맞음)
df_merged.loc[target_df.index, 'output_index'] = df_merged.loc[target_df.index, 'output_index'].combine_first(target_df['ref_output_index'])
df_merged.loc[target_df.index, 'labor_productivity_index'] = df_merged.loc[target_df.index, 'labor_productivity_index'].combine_first(target_df['ref_labor_productivity_index'])
```
